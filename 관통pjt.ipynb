# 관통 프로젝트

## 관통 프로젝트
- 매 주 금요일 All Day 프로젝트 
- 한 주 동안 배운 것을 모두 활용
- 단계적 협업을 하며 실무 경험을 쌓음
- 총 9번의 관통 + 최종 프로젝트(1주)

- 개발자로 가는 길
  - 알고리즘 + 프로젝트 + 컴퓨터 과학
  
  1. 알고리즘
    - 전략적인 접근
    - 자주 나오는 개념 학습
    - 난이도 별 문제풀이
    - 양을 늘려라
    - 코딩테스트 경험
  
  2. 프로젝트
    - 매력적인 나만의 서비스 만들기X
    - 기술을 활용할 수 있음O
    - 현업에서 원하는 기술 사용O
    - 혐업할 수 있는 인재O
    - 문제 해결 능력 어필O

  3. 컴퓨터 과학
    - '시나브로'의 힘
    - 혼자 X 같이 O
    - 스터디 적극 권장
    - 책 / 유튜브 / 블로그
    - 선택이 아닌 필수
